# Data mining \& clustering

The goal if this practical is to adress the folowing problem:
<center style="color:red" >  Given XXX raw, unlabeled documents, ... How to exploit/understand/represent them?</center>

In the previous week, we have seen how to represent textual data with the Bag of Words (BoW) model:
$$X =
	\begin{matrix}
	 & \textbf{t}_j \\
	 & \downarrow \\
	\textbf{d}_i \rightarrow &
	\begin{pmatrix}
	x_{1,1} & \dots & x_{1,d} \\
	\vdots & \ddots & \vdots \\
	x_{N,1} & \dots & x_{N,d} \\
	\end{pmatrix}
	\end{matrix}
	$$

From this BoW representation, we want to answer the following questions:
1. Which clustering algorithm to choose?
    - K-means, LSA, pLSA, LDA
1. What results to expect?
    - Semantics, noise cleaning, etc...
1. Which qualitative and quantitative analyses to understand the groups?

# 1) Clustering algorithm: K-Means

**Let's start by the most famous and simple unsupervised algorithm: $k$-means!**
Look at [sklear documentation](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)
and apply it to your BoW matrix.


```python
from sklearn.cluster import KMeans
# your code here
kmeans = KMeans(n_clusters=20, random_state=0, max_iter=10).fit(vectors) #vectors comprend 1000 mots par document
# Getting clusters:
centroides = kmeans.cluster_centers_
clusters = kmeans.labels_
print("Len de cluster=",len(clusters))
print(clusters)
print("Nb de centroides = nb de labels =",len(centroides))
print(centroides[0])
print(len(centroides[0]))
somme = 0
for c in centroides:
    somme += len(c)
print(somme) #on a bien mille points * 20 clusters

```

Chaque centroïde est un vecteur de 1000 dimensions, où chaque dimension représente la moyenne des occurrences d'un mot dans tous les documents du cluster.

Pour retrouver les mots dans un clusters : on multiplie le nième vecteur kmeans par le nieme vecteur de mots

```python
# TF-IDF
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=1000,stop_words='english')
vectors = vectorizer.fit_transform(newsgroups_train.data)

# KMeans
kmeans = KMeans(n_clusters=20, random_state=0, max_iter=10).fit(vectors) 

# Fonction pour obtenir les mots les plus importants par cluster (ici les 5 plus importants , n_words = 5)
def get_top_words_per_cluster(vectorizer, model, n_words=5): #le model c'est kmeans
    feature_names = vectorizer.get_feature_names_out() #on recupère les mots du vocabulaire
    clusters = {}
    for cluster_id in range(model.n_clusters):
        # Trouver les indices des mots les plus importants pour le cluster
        centroid = model.cluster_centers_[cluster_id]
        top_indices = centroid.argsort()[-n_words:][::-1] #Trie les indices des valeurs du centroïde par ordre croissant d'importance.
        top_words = [feature_names[i] for i in top_indices]
        clusters[cluster_id] = top_words
    return clusters

top_words = get_top_words_per_cluster(vectorizer, kmeans,3)
print("Mots les plus importants par cluster :", top_words)
```

```python
"""t-SNE (t-Distributed Stochastic Neighbor Embedding) est une technique de réduction de dimension qui permet de projeter 
des données haute-dimension (comme tes vecteurs de 1000 mots) dans un espace 2D ou 3D, 
tout en préservant au mieux les relations de voisinage entre les points."""


import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import matplotlib.cm as cm
#cm.get_cmap('jet', 20) : Crée une palette avec 20 couleurs distinctes (une par cluster).
cmap = cm.get_cmap('jet', 20)
# Réduction à 2D
tsne = TSNE(n_components=2, init='pca',max_iter=2500, verbose=2)
X_tsne = tsne.fit_transform(vectors.toarray())
# Visualisation avec t-SNE
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=clusters, cmap=cmap, s=10)
plt.title("Visualisation des clusters (t-SNE)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")

# Ajouter une légende pour les clusters
plt.colorbar(scatter, ticks=range(20), label="Cluster")
plt.show()
```

### Clustering Analysis:
1. **Qualtitative:**
    - Look at the most important words for each cluster
    - Perform cluster assignement to each document, and compute word cloud on the document (raw text or frequencies)
2. **Quantitative:**
    - Compute cluster "purity": $p_j= |y^*_j|$, where $y^*_j$ is the most frequent (GT) label in cluster $C_j$ $\Rightarrow$ $p = \frac{1}{N}\sum\limits_j  p_j$
    - Compute [Rand Score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.rand_score.html) and [Adjusted Rand Score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.adjusted_rand_score.html#sklearn.metrics.adjusted_rand_score)

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, rand_score, adjusted_rand_score

true_labels = Y #attibué dans le première cellule du notebook
clusters = kmeans.labels_

# Pureté 
def purity_score(true_labels, clusters):
    contingency_matrix = metrics.cluster.contingency_matrix(true_labels, clusters)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)

purity = purity_score(true_labels, clusters)
print(f"Pureté : {purity:.3f}")

# Rand Score et Adjusted Rand Score
rs = rand_score(true_labels, clusters)
ars = adjusted_rand_score(true_labels, clusters)
print(f"Rand Score : {rs:.3f}")
print(f"Adjusted Rand Score : {ars:.3f}")
```

# 2) Latent Semantic Analysis (LSA <=> SVD)


**Remember the LSA factorziation**:
$$
\begin{matrix}
 & X  &\!\!\!\!\!=\!\!\!\!\!& U  & \Sigma & V^T \\
  & \textbf{t}_j   &  & \hat{ \textbf{d}_i} & &  \\
 & \downarrow  &  &\downarrow  & & \\
\textbf{d}_i \rightarrow
&
\begin{pmatrix}
x_{1,1} & \dots & x_{1,d} \\
\\
\vdots & \ddots & \vdots \\
\\
x_{N,1} & \dots & x_{N,d} \\
\end{pmatrix}
&
\!\!\!\!\!=\!\!\!\!\!
%&
%(\hat{ \textbf{t}_j}) \rightarrow
&
\begin{pmatrix}
\begin{pmatrix} &  \textbf{u}_1 &  \end{pmatrix} \\
\vdots \\
\begin{pmatrix}  & \textbf{u}_k &  \end{pmatrix}
\end{pmatrix}
%&
%\!\!\!\!\!\cdot\!\!\!\!\!
&
\begin{pmatrix}
\sigma_1 & \dots & 0 \\
\vdots & \ddots & \vdots \\
0 & \dots & \sigma_k \\
\end{pmatrix}
%&
%\!\!\!\!\!\cdot\!\!\!\!\!
&
\begin{pmatrix}
\begin{pmatrix} \, \\ \, \\ \textbf{v}_1 \\ \, \\ \,\end{pmatrix}
\dots
\begin{pmatrix} \, \\ \, \\ \textbf{v}_k \\ \, \\ \, \end{pmatrix}
\end{pmatrix}
\end{matrix}
$$

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=20, n_iter=7, random_state=42)
U_svd = svd.fit_transform(vectors)  # Ici, U_svd ≈ U × Sigma (car TruncatedSVD renvoie U × Sigma)
print("U × Sigma (approximation) :\n", U_svd)
# Récupérer les valeurs singulières
Sigma = svd.singular_values_

# Normaliser chaque colonne de U_svd par la valeur singulière correspondante
U = U_svd / Sigma[np.newaxis, :]

print("Matrice U (vecteurs singuliers gauches) :\n", U)

print("dimensions de U:",U.shape)
print("dimensions de Sigma:",Sigma.shape)
```

- Look at [SVD doc in skelarn](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html#sklearn.decomposition.TruncatedSVD)
- Do the same qualitative/quantitative evaluation than with K-Means
- You can also use LSA as a pre-processing step for K-Means, *i.e.* running K-Means on $\boldsymbol{U}$ matrix above
    - N.B. : try without/with $\ell_2$ normalization of $\boldsymbol{U}$'s rows before running  K-Means
    - You can also benefit from LSA pre-processing for using [t-SNE visualization](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html) (see code below)


TNSE : Intégration stochastique de voisins à distribution T.
Traitement des données statistiques, par le calcul des probabilités.
