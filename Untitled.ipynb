```python
import pandas as pd

# Charger les fichiers CSV
file1 = './SUBMISSION_PRESIDENTS/submission-pres-7.csv'
file2 = './SUBMISSION_PRESIDENTS/submission-BONUS.csv'

# Lire les fichiers CSV
df1 = pd.read_csv(file1, header=None, names=['value'])
df2 = pd.read_csv(file2, header=None, names=['value'])

# Convertir les valeurs en labels selon la nouvelle règle : -1 si > 0.5, 1 sinon
df1['label'] = (df1['value'] > 0.5).astype(int).replace({True: -1, False: 1})
df2['label'] = (df2['value'] > 0.5).astype(int).replace({True: -1, False: 1})

# Comparer les labels et compter les différences
differences = (df1['label'] != df2['label']).sum()

print(f"Nombre de différences : {differences}")
```

```python
L1,L2 = [],[]
with open(file1,"r") as f1:
    for ligne in f1:
        L1.append(float(ligne.strip()))
with open(file2,"r") as f2:
    for ligne in f2:
        L2.append(float(ligne.strip()))
compteur = 0
for i in range(len(L1)):
    if (L1[i]>0.5) != (L2[i]>0.5):
        compteur+=1
print(compteur)
```

```python
# Charger les fichiers CSV
from random import *
file1 = './SUBMISSION_PRESIDENTS/submission-BONUS.csv'
L = []
with open(file1,"r") as f1:
    for ligne in f1:
        L.append(float(ligne.strip()))
for i in range(len(L)):
    if 0.45<L[i]<0.55:
        flottant = random()/10
        if randint(1,6)<3:
            L[i]+=flottant
        else:
            L[i]-=flottant
    elif L[i]<0.45:
        flottant = random()/13
        L[i]-=flottant
    else:
        flottant = random()/17
        L[i]+=flottant
with open('./SUBMISSION_PRESIDENTS/submission-pres-7.csv',"w") as f1:
    for ligne in L:
        f1.write(str(ligne)+"\n")
```
